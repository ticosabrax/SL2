## K-means

```python
# Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargando datos
data = pd.read_csv("data/estaturas.csv", sep=";").to_numpy()
data
```

```python
# Plot de datos
plt.scatter(data[:,0], data[:,1])
plt.title("Para Clustering")
plt.ylabel("Edad")
plt.xlabel("Estatura")
plt.show()
```

### Implementando K-means

```python
class Kmeans:

    # Inicializando clase
    def __init__(self, k=1):
        self.k = k
        self.shape = None
        self.ranges = None
        self.centroids = None
        self.distances = None
        self.labels = None
        self.minDistances = None
        self.error = None
        
    # Configurando variables internas para el algoritmo    
    def setConfig(self, data):
        self.shape = data.shape
        
        # Cálculo de rangos para centroides
        self.ranges = np.zeros((self.shape[1], 2))
        for dim in range(self.shape[1]):
            self.ranges[dim, 0] = np.min(data[:,dim])
            self.ranges[dim, 1] = np.max(data[:,dim])

        # Cálculo de centroides para k clases
        self.centroids = np.zeros((self.k, self.shape[1]))
        for i in range(self.k):
            for dim in range(self.shape[1]):
                self.centroids[i, dim] = np.random.uniform(self.ranges[dim, 0], self.ranges[dim, 1], 1)
 
        plt.scatter(data[:,0], data[:,1])
        plt.scatter(self.centroids[:,0], self.centroids[:,1], c = 'r')
        plt.title("Centroides")
        plt.ylabel("Edad")
        plt.xlabel("Estatura")
        plt.show()
        print('=================================================')
            
    # Modelo de entrenamiento
    def fit(self, data, verbose=1, graph=1):
        
        self.setConfig(data)
        lastCentroids = np.zeros((self.k, self.shape[1]))
        count = 0
        
        while not (self.centroids == lastCentroids).all():
            count+=1
            if verbose == 1:
                print('Iteración: ', count)
        
            self.distances = np.zeros((self.shape[0],self.k))
            for ix, i in enumerate(data):
                for ic, c in enumerate(self.centroids):
                    self.distances[ix, ic] = np.sqrt(np.sum((i-c)**2))

            labels = np.argmin(self.distances, axis = 1)

            new_centroids = np.zeros((self.k, self.shape[1]))
            for centroid in range(self.k):
                temp = data[labels == centroid]
                if len(temp) == 0:
                    return 0
                for dim in range(self.shape[1]): 
                    new_centroids[centroid, dim] = np.mean(temp[:,dim])

            if graph == 1:
                plt.scatter(data[:,0], data[:,1], c = labels)
                plt.scatter(new_centroids[:,0], new_centroids[:,1], c = 'r')
                plt.show()

            lastCentroids = self.centroids
            self.centroids = new_centroids
            self.labels = labels
            self.Error()
        
            if verbose == 1:
                print('Centroides')
                print(self.centroids)
                print('Distortion error')
                print(self.error)
                print('=================================================')
        return 1
    
    # Distancias mínimas
    def setMinDistances(self):
        self.minDistances = np.zeros((self.shape[0]))
        for index in range(self.shape[0]):
            self.minDistances[index] = self.distances[index, self.labels[index]]

    # Cálculo de error
    def Error(self):
        self.setMinDistances()
        self.error = np.mean(self.minDistances)
    
    def predict(self, data, graph=1):
        distances = np.zeros((data.shape[0],self.k))
        for ix, i in enumerate(data):
            for ic, c in enumerate(self.centroids):
                distances[ix, ic] = np.sqrt(np.sum((i-c)**2))
        
        labels = np.argmin(distances, axis = 1)
        
        if graph == 1:
            plt.scatter(data[:,0], data[:,1], c = labels)
            plt.scatter(self.centroids[:,0], self.centroids[:,1], c = 'r')
            plt.show()
        return labels
```

### Realizando pruebas con K-means

```python
kmeans = Kmeans(5)
kmeans.fit(data)
```

```python
kmeans.labels
```

```python
kmeans.predict(data)
```

## GMM

```python
# Librerías
from sklearn.mixture import GaussianMixture

# Entrenando modelo
gmm = GaussianMixture(n_components = 5)
gmm.fit(data)
gmm_predict = gmm.predict(data)
```

```python
# Plot
plt.scatter(data[:,0], data[:,1], c = gmm_predict)
plt.show()
```

```python
# Score
gmm.score(data)
```

## Seleccionando K

```python
np.random.seed(95)
errores = list()
models = list()

for i in range(2,7):
    print(i)
    model = Kmeans(i)
    model.fit(data, verbose=0, graph=0)
    models.append(model)
    errores.append(model.error)
    
errores
```

### Gráfica de codo

```python
# Gráfica de codo
plt.plot(np.array([2,3,4,5,6], dtype = np.int32), errores)
plt.scatter(np.array([2,3,4,5,6], dtype = np.int32), errores)
plt.title("K means Distortion error")
plt.xlabel("K")
plt.ylabel("Distortion error")
plt.xticks([2,3,4,5,6])
plt.show()
```

> Con base a la gráfica se observa que cuando **K=3** se genera el codo, por lo cual el **K** a utilizar es 3.

### Calculando modelos con k=3

#### Kmeans

```python
modelKmeans = models[1]
modelKmeans.fit(data)
```

#### GMM

```python
np.random.seed(95)
modelGMM = GaussianMixture(n_components=3)
modelGMM.fit(data)
gmm_predict = modelGMM.predict(data)
```

```python
plt.scatter(data[:,0], data[:,1], c=gmm_predict)
plt.title("GMM ")
plt.ylabel("Edad")
plt.xlabel("Estatura")
plt.show()
```

### Análisis y resultados

Al observar los resultados, los algoritmos nos indican que los datos los podemos agrupar en 3 grupos (clusters), así mismo los grupos están fuertemente influenciados o definidos por la edad, podemos ver que un grupo está comprendido entre los 20 a 29 años aproximandamente, el otro grupo entre 30 a 40 años y el tercer grupo de 41 y más años. La estatura pareciera que no aporta algún efecto en la agrupación.

A manera de conclusión podría indicar que los modelos agrupan de manera adecuada los datos, tomando en consideración que la Edad es la variable que genera un mayor efecto en la dicha agrupación.

### Función $f(edad, estatura) = estatura /edad$

```python
def g(X):
    estatura = X[:,0]
    edad = X[:,1]
    return estatura/edad
```

### Calculando valor esperado

```python
muestra_g = g(muestra)
valor_esperado_g = np.mean(muestra_g)
valor_esperado_g
```

```python
# Graficando
plt.hist(muestra_g, density=True)
plt.axvline(valor_esperado_g,color="k",linestyle="--")
plt.title("El valor esperado de g(X) es:{}".format(valor_esperado_g))
plt.show()
```
